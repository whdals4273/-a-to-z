```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('DATA_SF_1000S_SAMPLING.csv', sep = ',')

df.head(10)


```

```python
fig, ax  = plt.subplots()
df1 = df.sort_values('time')

group = df1.groupby('VREF')


for VREF, g in group:
    
    plt.plot(g['time'],g['VOUT'],marker = 'o',markerfacecolor = 'none',label = f"VREF = {VREF}[V]")

ax.set_xlabel('time[s]')
ax.set_ylabel('Vout[V]')
ax.set_title('source follower_sampling', fontsize = 16)
ax.legend(title = 'Vref[V]',bbox_to_anchor=(1,1))

ax.grid(True)
```
