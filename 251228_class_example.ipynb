```python
class Car:
    def __init__(self, model, fuel=50):                             # car라는 class를 형성할 때 기본적으로 가지고 있을 속성을 정하는 것
        # 생성될 때(초기화) 각 차가 가져야 하는 상태(속성)
        self.model = model
        self.fuel = fuel
        self.speed = 0

    def accelerate(self, amount=10):                                # 만들어진 car class에 적용하는 함수를 정의한 것이고 실행하려면 .accelerate() 형태로 해야함
        # 행동(메서드): 속도 증가, 연료 감소
        if self.fuel <= 0:
            return "연료 없음"
        self.speed += amount
        self.fuel -= 1
        return f"{self.model}: speed={self.speed}, fuel={self.fuel}"

    def brake(self, amount=10):
        self.speed = max(0, self.speed - amount)
        return f"{self.model}: speed={self.speed}"

    def refuel(self, amount):
        self.fuel += amount
        return f"{self.model}: fuel={self.fuel}"
```

```python
my_car = Car("Avante",fuel = 5)

```

```python
print(my_car.refuel(10000))
```

```python
my_car.accelerate()
print(my_car.speed,my_car.fuel)
print(my_car.model)
```
