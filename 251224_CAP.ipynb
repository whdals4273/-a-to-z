```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('[251203_#1_cap100]C-V with freq 1st [-3.00;0.10;3.00]~[0.03,CPD]-0.txt', sep = '\t')

row_count = df.shape[0]

print(df)
```

```python
plt.figure(figsize=(7, 5))

x = df['V']

for col in df.columns[1:]:   # V   
    plt.plot(x, df[col], marker='o', markersize=3, label=col)

plt.xlabel('Voltage[V]')
plt.ylabel('capacitance[F]')
plt.legend(title='Frequency',bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True)
plt.show()
```
