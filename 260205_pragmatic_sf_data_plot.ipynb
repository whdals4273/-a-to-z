```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Vin_Vout_4.csv', sep = ',')

df.head(10)


```

```python
fig, ax = plt.subplots()
df1 = df.sort_values(['VREF','VIN'])
group = df1.groupby('VREF')
for i,g in group:
    plt.plot(g['VIN'],g['VOUT'],marker = 'o', markerfacecolor = 'none',label = f'Vref ={i}[V]')

ax.set_xlabel('Vin[V]')
ax.set_ylabel('Vout[V]')
ax.grid(True)
ax.set_title('source follower_pragmatic', fontsize = 16)
ax.legend(bbox_to_anchor = (1,1))
```
