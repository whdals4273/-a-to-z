```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 251111  wafer

raw_data = pd.read_csv('I_V List Sweep - Step Marking (ISRC) [IGZO_ALD_#1_2T1C_LIST_SWEEP_R0C1_VD_PGM_CONST(4) ; 12_24_2025 2_40_29 PM].csv', sep = ',',skiprows = 251)

df = raw_data.iloc[:,1:]
row_count = df.shape[0]
df.columns = df.columns.str.strip()   
  
plt.figure()
for row in range(row_count):
    plt.semilogy(df["time"],df["ID_READ"], marker = 'o', markerfacecolor = 'none',markersize=3)
plt.grid(True, which = 'both')
plt.xlabel('time[s]')
plt.ylabel('Id_read[A]')

plt.figure()
fig, axes = plt.subplots(2, 1, sharex=True)

axes[0].plot(df['time'], df['VG_PGM'],marker = 'o',  markerfacecolor = 'none',markersize=3)
axes[0].set_ylabel('VG_PGM')
axes[0].grid(True, which='both')

axes[1].plot(df['time'], df['VD_PGM'],marker = 'o', markerfacecolor = 'none',markersize=3)
axes[1].set_xlabel('Time(s)')
axes[1].set_ylabel('VD_PGM')
axes[1].grid(True)
    
```
