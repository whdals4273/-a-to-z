```python
import pandas as pd

file_name = 'datasets/German_credit/German_credit.csv'
df = pd.read_csv(file_name, sep = ',')
```

```python
column_name = df.columns

for i in column_name:
    empty = df[i].isna().sum()
    print(f'{i} 컬럼의 결측값 갯수는 {empty} 개 입니다.')
```

```python
import numpy as np

df_ex = pd.DataFrame({'time' : [1,2,5,6,7],'value' : [0,2,5,6,8]})

df_ex.iloc[1,1] = np.nan
print(df_ex)
```

```python
df_ex  = df_ex.set_index('time')

df_ex = df_ex.interpolate(method = 'index', order = 1, limit_direction='both',limit = 2)
print(df_ex)
```

```python
df_ex = df_ex.reset_index()
```

```python
target = 2.5

df_new = pd.concat([df_ex, pd.DataFrame({'time': [target]})])

print(df_new)

df_new = df_new.set_index('time')
df_new = df_new.interpolate(method = 'index')
df_new = df_new.reset_index()

target_value = df_new.loc[df_new['time'] ==2.5,'value'].iloc[0]
print(target_value)
```

```python
print(f'2.5초일때 value는 {target_value}')
```
